```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_excel("zomato data.xlsx",sheet_name=None)
```

```python
# Get the names of all sheets
sheet_names=df.keys()

a=list(sheet_names)
a
```

```python
for i in range(len(a)):
    a[i]=pd.read_excel("zomato data.xlsx",sheet_name=a[i])
```

```python
users=a[0]
menu=a[1]
restaurant=a[2]
orders=a[3]
food=a[4]
```

```python
users
```

```python
#Data Cleaning for User sheet

users.isnull().sum()       #As u can see we did not need to clean data          
```

```python
menu
```

```python
#Data Cleaning for Menu sheet

menu.isnull().sum() #we Have one null value so we can drop it
```

```python
menu=menu.dropna()
menu.isnull().sum()
```

```python
restaurant
```

```python
#Data Cleaning for Menu sheet

(restaurant.isnull().sum()/restaurant.shape[0])*100     #as we can see that the % of null values w.r.t total rows is max to max 0.6 % so we can drop them 
```

```python
restaurant=restaurant.dropna()
restaurant.isnull().sum()
```

```python
orders
```

```python
#Data Cleaning for ordes sheet

orders.isnull().sum()/orders.shape[0]  #as we can see that the % of null values w.r.t total rows is 0.01 % so we can drop them 
```

```python
orders=orders.dropna()
orders.isna().sum()
```

```python
food
```

```python
#Data Cleaning for ordes sheet
food.isnull().sum()
```

```python
with pd.ExcelWriter("zomato data.xlsx", engine="openpyxl", mode="a", if_sheet_exists="replace") as writer:
    users.to_excel(writer, sheet_name="users", index=False, header=True)
    menu.to_excel(writer, sheet_name="menu", index=False, header=True)
    restaurant.to_excel(writer, sheet_name="restaurant", index=False, header=True)
    orders.to_excel(writer, sheet_name="orders", index=False, header=True)
    food.to_excel(writer, sheet_name="food", index=False, header=True)
```
